Create standard and multi-series bar plots. Bar graphs are used to compare data between different groups or to track changes over time (similar to a line chart). Multi-series Bar graphs can display multiple groups in the same bar in a stacked visual showing the composition of the bar chart along the Y-axis.

Matplotlib provides the bar() function for creating bar plots.

Use the bar() function parameters below with the axes object:

Given the data y_pos and height below, use the function bar() to create a bar plot. Set the argument color inside the function bar() equal to (0.2, 0.4, 0.6, 0.6). Make these customizations to the plot:

Title: "Bar plot"
X-ticks: Use (y_pos, bars) as arguments for plt.xticks.
Save your plot as a png with the name "plot2.png".

```python
# Data
height = [4, 12, 5, 16, 45]
bars = ('A', 'B', 'C', 'D', 'E')
y_pos = np.arange(len(bars))

### BEGIN SOLUTION
plt.bar(y_pos, height, color=(0.2, 0.4, 0.6, 0.6))

# title
plt.title("Bar plot")


plt.xticks(y_pos, bars)


plt.savefig("plot2.png")


plt.show()

### END SOLUTION
```

## Question 6

Using the patients.csv dataframe df, create a univariate bar graph using the 'age' dimension from the df dataframe. Ensure that the bar plot display count is only 10 values.

10 points

```python
### BEGIN SOLUTION
age_counts = df['age'].value_counts().nlargest(10)
plt.figure(figsize=(7, 5))
age_counts.plot(kind='bar', color='green')

plt.title('Top 10 Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')

plt.xticks(rotation=45)
plt.show()
### END SOLUTION
```
